# Split Data

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.3, random_state=4)
```

```python
X_train.shape
```

```python
import umap
import umap.plot
```

```python
reducer = umap.UMAP(n_neighbors = 1000,min_dist=0.30,n_components=1)
```

```python
reducer.fit(X_train)
```

```python
reducer2D = umap.UMAP(n_neighbors = 1000,min_dist=0.30,n_components=2)
```

```python
reducer2D.fit(X_train)
```

```python
umap.plot.points(reducer2D,labels=y_train)
```

```python
umap.plot.points(reducer2D,theme="darkgreen")
```

```python
umap.plot.points(reducer2D,y_train, theme="fire")
```

```python
umap.plot.points(reducer2D,y_train,color_key_cmap='Paired', theme="darkgreen")
```

```python
reducer2D_1 = umap.UMAP(n_neighbors = 100,min_dist=0.25,n_components=2)
```

```python
reducer2D_1.fit(X_train)
```

```python
umap.plot.points(reducer2D_1,y_train,color_key_cmap='Paired', theme="darkgreen")
```

```python
umap.plot.points(reducer2D_1)
```

```python
umap.plot.points(reducer2D_1, theme="darkgreen")
```
